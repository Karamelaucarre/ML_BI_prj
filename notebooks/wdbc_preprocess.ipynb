```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
import os
save_path = '../data/preprocessed'
os.makedirs(save_path, exist_ok= True)
```

```python
dataset = pd.read_csv('../data/original/wdbc.data', sep=',', header=None)
```

```python
cols = []
for i in range(32):
    if i == 1:
        cols.append('Target')
    elif i ==0:
        cols.append('ID')
    else:
        cols.append(f'feature{i-1}')
len(cols)

target_name = 'Target'
```

```python
dataset.columns = cols
dataset
```

```python
dataset.isnull().sum() #num of null values
```

```python
dataset.drop('ID', axis=1, inplace=True) #Drop the column
```

```python
dataset
```

```python
# 1. Créer la liste des colonnes X (toutes les colonnes sauf la cible)
colonnes_X = [col for col in dataset.columns if col != 'Target']

# 2. Créer le nouvel ordre : X, puis y
nouvel_ordre = colonnes_X + ['Target']

# 3. Réordonner le DataFrame
dataset = dataset[nouvel_ordre].copy()
```

```python
dataset.to_csv(f'../data/original/wdbc.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset = np.array(dataset)
dataset[:, -1] = le.fit_transform(dataset[:, -1])
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
dataset[:, :-1] = st.fit_transform(dataset[:, :-1])
```

```python
dataset= pd.DataFrame(dataset)
dataset
```

```python
dataset.to_csv(f'{save_path}/wdbc.csv')
```
